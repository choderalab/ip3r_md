```python
import mdtraj as md 
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Load trajecories

adenosine_traj=md.load_dcd('path/to/joined.dcd',top='path/to/step3_input.pdb')
guanosine_traj=md.load_dcd('path/to/joined.dcd',top='path/to/step3_input.pdb')
```

```python
# Define function to find distance between two atoms


def atom_distance(traj, atom1_index, atom2_index):

    atom_pairs = [[atom1_index, atom2_index]]
    distances = md.compute_distances(traj, atom_pairs)
    return distances[:, 0]
```

```python
# Find the distance between adenosine C5 and Cys 2562 alpha-carbon

adenosine_dist=atom_distance(adenosine_traj,6371,5119)
```

```python
# Find the distance between guanosine C5 and Cys 2562 alpha-carbon

gsn_dist=atom_distance(guanosine_traj,6382,5119)
```

```python
# Plot the distances

plt.figure(figsize=(4, 3))
plt.plot(adenosine_dist[0:150]*10,linewidth=4,color='crimson')
plt.plot(gsn_dist[1001:1150]*10,linewidth=4,color='royalblue')
plt.xlim(0,150)
plt.savefig('asn_vs_gsn.png',dpi=300)
```
